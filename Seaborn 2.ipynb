```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sb
```

```python
dtst=pd.read_csv('placement.csv')
```

```python
sb.boxplot(x='ssc_b',y='ssc_p',data=dtst,hue='gender')
plt.show()
```

```python
sb.barplot(x='ssc_b',y='ssc_p',data=dtst,hue='gender')
plt.show()
```

```python
sb.countplot(x='ssc_b',data=dtst,hue='gender')
plt.show()
```

```python
sb.pointplot(x='ssc_p',y='hsc_b',data=dtst,hue='gender')
plt.show()
```

```python
sb.boxplot(data=dtst,orient='h')
plt.show()
```

```python
import seaborn as sb
from matplotlib import pyplot as plot
df=sb.load_dataset('exercise')
sb.catplot(x='time',y='pulse',hue='kind',kind='violin',col='diet',data=df);
plt.show()
```

```python
df=sb.load_dataset('tips')
sb.regplot(x='total_bill',y='tip',data=df)
sb.lmplot(x='total_bill',y='tip',data=df)
plt.show()
```

```python
df=sb.load_dataset('anscombe')
sb.lmplot(x='x',y='y',data=df.query("dataset=='II'"))
plt.show()
```

```python
df=sb.load_dataset('anscombe')
sb.lmplot(x='x',y='y',data=df.query("dataset=='II'"),order=2)
plt.show()
```

```python
df=sb.load_dataset('tips')
g=sb.FacetGrid(df,col="time")
g.map(plt.hist,"tip")
plt.show()
```

```python
df=sb.load_dataset('iris')
g=sb.PairGrid(df)
g.map(plt.scatter)
plt.show()
```
